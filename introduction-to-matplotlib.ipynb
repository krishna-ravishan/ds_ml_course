# Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plottting in Jypyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x, y)

#4. Customize plot 
ax.set(title="Simple plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and show (You save the whole figure)
fig.savefig("sample-plot.png")
```

# Two options for subplots

```python
# Subplot option 1 (Recommended)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1]. hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib OO Method?)
* When plotting something quickly, it is better to use pyplot method
* But when plotting something more advanced, use the OO method


```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#Pyplot Method
over_50.plot(kind = "scatter", 
            x="age",
            y="chol",
            c="target")
```

```python
# OO Method mixed with the pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
# ax.set_xlim([45, 100])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease & Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```

```python
over_50.head()
```

```python
# Subplot of chol age thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               figsize=(10, 10))

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                     c=over_50["target"])
# Customize ax0
ax0.set(title="Heart Disease & Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

# Now doing the same for ax1
scatters = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"]);

ax1.set(title="Heart Disease & Max Heart Rate",
        xlabel="Age",
        ylabel="Thalach");

ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

# Title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
